## Auto Regressive

## An autoregressive (AR) model forecasts future behavior based on past behavior data. This type of analysis is used when there is a correlation between the time series values and their preceding and succeeding values.

```python


class plots:
    def __init__(self, data):
        self.data = data

    def hourly_plot(self, hours):

      for hour in hours:
        dates=(self.data)['Date']
        series=self.data['F'+hour] if hour!='AVERAGE LOAD' else self.data[hour]

        plt.figure(figsize=(10, 6))
        plt.plot(dates, series)
        plt.title(f'Hourly Plot for Hour {hour}')
        plt.xlabel('Date')
        plt.ylabel('Value')
        plt.grid(True)
        plt.tight_layout()
      plt.show()
```

```python
visualize=plots(raw_df)
visualize.hourly_plot([str(x) for x in range(3,27)])
```

```python
visualize.hourly_plot(['AVERAGE LOAD'])
```

# not a stationary

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

av = average_df['AVERAGE LOAD']

fig, axes = plt.subplots(1, 2, sharex=True, figsize=(12, 6))


axes[0].plot(av, label='Original Series')
axes[0].plot(av.rolling(window=10).mean(), color='r', label='Rolling Mean')
axes[0].set_title('Original Series with Rolling Mean')
axes[0].legend()

plot_acf(av, ax=axes[1])
axes[1].set_title('Autocorrelation Function')

plt.show()
```

```python
shift=3.1416*4
test_x=np.arange(0,3.1416*4,0.1)
test_y=np.sin(test_x)

fig,ax=plt.subplots(1,2)
ax[0].plot(test_x,test_y,c='r')
ax[0].plot(test_x+shift,np.sin(test_x),c='b')

from statsmodels.graphics.tsaplots import plot_acf
plot_acf(test_y,lags=shift,ax=ax[1])


```

```python
def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )

    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
    else:
        print("weak evidence against null hypothesis,indicating it is non-stationary ")

adfuller_test(av)
```
